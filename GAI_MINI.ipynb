```python
import cv2
import cv2
from matplotlib import pyplot as plt

def cartoonize_anime_style(img_path):
    # Read the image
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Step 1: Apply bilateral filter to smooth while preserving edges
    smooth = cv2.bilateralFilter(img_rgb, d=9, sigmaColor=300, sigmaSpace=300)

    # Step 2: Convert to grayscale and apply median blur
    gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
    blur = cv2.medianBlur(gray, 7)

    # Step 3: Detect edges using adaptive thresholding
    edges = cv2.adaptiveThreshold(blur, 255,
                                   cv2.ADAPTIVE_THRESH_MEAN_C,
                                   cv2.THRESH_BINARY, blockSize=9, C=2)

    # Step 4: Combine edges with the smoothed image
    cartoon = cv2.bitwise_and(smooth, smooth, mask=edges)

    # Step 5: Display result
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.title("Original Image")
    plt.imshow(img_rgb)
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.title("Anime Style Cartoonized")
    plt.imshow(cartoon)
    plt.axis('off')
    plt.show()

# Replace 'face.jpg' with your image path
cartoonize_anime_style("/content/dhoni.jpg")

# The following code block was unintentionally indented and has been removed
#img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Step 1: Apply bilateral filter to smooth while preserving edges
#smooth = cv2.bilateralFilter(img_rgb, d=9, sigmaColor=300, sigmaSpace=300)

# Step 2: Convert to grayscale and apply median blur
#gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
#blur = cv2.medianBlur(gray, 7)

# Step 3: Detect edges using adaptive thresholding
#edges = cv2.adaptiveThreshold(blur, 255,
                                   #cv2.ADAPTIVE_THRESH_MEAN_C,
                                 #  cv2.THRESH_BINARY, blockSize=9, C=2)

# Step 4: Combine edges with the smoothed image
#cartoon = cv2.bitwise_and(smooth, smooth, mask=edges)

# Step 5: Display result
#plt.figure(figsize=(10,5))
#plt.subplot(1,2,1)
#plt.title("Original Image")
#plt.imshow(img_rgb)
#plt.axis('off')

#plt.subplot(1,2,2)
#plt.title("Anime Style Cartoonized")
#plt.imshow(cartoon)
#plt.axis('off')
#plt.show()

# Replace 'face.jpg' with your image path
#cartoonize_anime_style("/content/SAI.1.jpg")
```
